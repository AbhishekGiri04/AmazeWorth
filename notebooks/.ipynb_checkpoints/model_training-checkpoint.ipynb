# ðŸ“˜ Amazon ML Challenge - Model Training Notebook

# Step 1 â€” Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.sparse
from tqdm import tqdm
import joblib
import pickle
import os
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from scipy.sparse import hstack, csr_matrix
```

# Step 2 â€” Configuration

```python
DATA_DIR = '..backend/data/'
FEATURE_DIR = '..backend/image_features/'
MODEL_DIR = '..backend/models/'

MODEL_VECTOR_SIZE = 1280
RANDOM_SEED = 42

print("Directory structure verified!")
print(f"DATA_DIR: {DATA_DIR}")
print(f"FEATURE_DIR: {FEATURE_DIR}")
print(f"MODEL_DIR: {MODEL_DIR}")
```

# Step 3 â€” Helper Functions

```python
def smape(y_true, y_pred):
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    diff = np.abs(y_true - y_pred) / denominator
    diff[denominator == 0] = 0.0
    return 100 * np.mean(diff)
    
```

# Step 4 â€” Load and Explore Data

```python
train = pd.read_csv("../backend/data/train.csv")
test = pd.read_csv("../backend/data/test.csv")

print("Train & Test Data Loaded")
print(f"Train shape: {train.shape}")
print(f"Test shape: {test.shape}")
display(train.head())

print(f"Training samples: {len(train)}")
print(train.describe())

sns.histplot(train['price'], bins=50, kde=True)
plt.title("Price Distribution (Before Log Transform)")
plt.show()
```

# Step 5 â€” Prepare Features

```python
print("Creating text + brand + numeric + image features...")

# TEXT FEATURES
text_col = "catalog_content"
train[text_col] = (
    train[text_col]
    .fillna("")
    .str.lower()
    .str.replace(r"[^a-z0-9 ]", " ", regex=True)
)

tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1, 3), min_df=3)
X_text = tfidf.fit_transform(train[text_col])
print(f"TF-IDF text feature shape: {X_text.shape}")

# BRAND FEATURE
if "brand" in train.columns:
    label_encoder = LabelEncoder()
    X_brand = label_encoder.fit_transform(train["brand"].astype(str)).reshape(-1, 1)
    print("Brand feature encoded successfully.")
else:
    print("No 'brand' column found â€” using zeros instead.")
    X_brand = np.zeros((len(train), 1))

# NUMERIC FEATURES
train["text_len"] = train[text_col].apply(len)
train["word_count"] = train[text_col].apply(lambda x: len(x.split()))
numeric_cols = ["text_len", "word_count"]
X_num = train[numeric_cols].values
print(f"Numeric features shape: {X_num.shape}")

# IMAGE FEATURES
image_features = []
for sid in tqdm(train["sample_id"], desc="Loading image features"):
    path = f"{FEATURE_DIR}{sid}.npy"
    if os.path.exists(path):
        image_features.append(np.load(path))
    else:
        image_features.append(np.zeros(MODEL_VECTOR_SIZE))

X_img = np.array(image_features)
print(f"Loaded raw image features: {X_img.shape}")

# PCA REDUCTION
if X_img.shape[1] > 128:
    print(f"Reducing image feature size from {X_img.shape[1]} â†’ 128 using PCA...")
    X_img = np.nan_to_num(X_img)  # prevent NaN warnings
    pca = PCA(n_components=128, random_state=42, svd_solver='full')
    X_img = pca.fit_transform(X_img)
    print(f"Reduced image feature shape: {X_img.shape}")

# FINAL FEATURE STACK 
X = hstack([X_text, X_brand, X_num, X_img], format="csr")
print(f"Final feature matrix shape: {X.shape}")
```

# Step 6 â€” Split Data

```python
y = train["price"].values
y_log = np.log1p(y) 

X_train, X_val, y_train, y_val = train_test_split(
    X, y_log, test_size=0.15, random_state=RANDOM_SEED
)
print(f"Training on {X_train.shape[0]} samples, validating on {X_val.shape[0]} samples.")
```

# Step 7 â€” Train Model

```python
params = {
    'objective': 'regression_l1',
    'metric': 'rmse',
    'learning_rate': 0.05,
    'n_estimators': 1000,    
    'num_leaves': 63,      
    'colsample_bytree': 0.7,
    'subsample': 0.7,
    'max_bin': 255,
    'min_data_in_leaf': 50,
    'reg_alpha': 0.1,
    'reg_lambda': 0.1,
    'force_col_wise': True,   
    'n_jobs': 2,
    'random_state': RANDOM_SEED
}
print("Training LightGBM model with early stopping...")

model = lgb.LGBMRegressor(**params)

model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    eval_metric='rmse',
    callbacks=[lgb.early_stopping(50), lgb.log_evaluation(100)]
)

print("Best iteration:", model.best_iteration_)
```

# Step 8 â€” Evaluate Model

```python
y_val_pred = np.expm1(model.predict(X_val))
y_val_true = np.expm1(y_val)

rmse = np.sqrt(mean_squared_error(y_val_true, y_val_pred))
smape_score = smape(y_val_true, y_val_pred)

print(f"RMSE: {rmse:.2f}")
print(f"SMAPE: {smape_score:.2f}%")

plt.figure(figsize=(7,5))
sns.scatterplot(x=y_val_true, y=y_val_pred, alpha=0.5)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title(f"Predicted vs Actual Prices (SMAPE={smape_score:.2f}%)")
plt.show()
```

# Step 9 â€” Feature Importance

```python
lgb.plot_importance(model, max_num_features=20, importance_type='gain')
plt.title("Top 20 Feature Importances (LightGBM)")
plt.show()
```

# Step 10 â€” Save Artifacts

```python
# Make sure backend/models folder exists
MODEL_DIR = "../backend/models"

# Save trained LGBM model
joblib.dump(model, os.path.join(MODEL_DIR, "lgbm_final_model.pkl"))

# Save TF-IDF vectorizer
with open(os.path.join(MODEL_DIR, "tfidf_vectorizer.pkl"), "wb") as f:
    pickle.dump(tfidf, f)

print("Model and TF-IDF vectorizer saved successfully in backend/models/")
```

# Step 11 â€” Sample Prediction Check

```python
sample_idx = np.random.randint(0, len(y_val_true))
print("\nSample Prediction Check:")
print("Actual Price:", round(y_val_true[sample_idx], 2))
print("Predicted Price:", round(y_val_pred[sample_idx], 2))
```

# Step 12 â€” Distribution Comparison

```python
plt.figure(figsize=(8,4))
sns.kdeplot(y_val_true, label="Actual Prices", fill=True, color='blue')
sns.kdeplot(y_val_pred, label="Predicted Prices", fill=True, color='orange')
plt.title("Price Distribution: Actual vs Predicted")
plt.xlabel("Price")
plt.ylabel("Density")
plt.legend()
plt.show()
```

# Step 13 â€” Predict on Test Set

```python
import re

# Define paths
MODEL_DIR = os.path.join("..", "backend", "models")
DATA_DIR = os.path.join("..", "backend", "data")

# Load model and TF-IDF
print("Loading trained model and TF-IDF vectorizer...")
final_model = joblib.load(os.path.join(MODEL_DIR, "lgbm_final_model.pkl"))
with open(os.path.join(MODEL_DIR, "tfidf_vectorizer.pkl"), "rb") as f:
    tfidf = pickle.load(f)
print("Model and TF-IDF vectorizer loaded successfully!")

# Load test.csv
test_path = os.path.join(DATA_DIR, "test.csv")
test = pd.read_csv(test_path)
print(f"Loaded test.csv with shape: {test.shape}")

# Clean text helper
def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r"[^a-z0-9 ]", " ", text)
    return text.strip()

# Clean text column
print("Cleaning test data catalog_content...")
test["catalog_clean"] = test["catalog_content"].apply(clean_text)

# Create numeric features
test["text_len"] = test["catalog_clean"].apply(len)
test["word_count"] = test["catalog_clean"].apply(lambda x: len(x.split()))
numeric_cols = ["text_len", "word_count"]

# Prepare feature sets
print("Generating features for test.csv...")
X_text_test = tfidf.transform(test["catalog_clean"])

if "brand" in test.columns:
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    X_brand_test = le.fit_transform(test["brand"].astype(str)).reshape(-1, 1)
else:
    X_brand_test = np.zeros((test.shape[0], 1))

X_num_test = test[numeric_cols].values

# Combine features
X_test = hstack([
    X_text_test,
    csr_matrix(X_brand_test),
    csr_matrix(X_num_test)
], format='csr')

# Fix feature mismatch (padding/trimming)
expected_features = final_model.n_features_in_
current_features = X_test.shape[1]

if current_features != expected_features:
    diff = expected_features - current_features
    print(f"Feature mismatch: model expects {expected_features}, got {current_features}")
    if diff > 0:
        # Add zero padding columns
        padding = csr_matrix(np.zeros((X_test.shape[0], diff)))
        X_test = hstack([X_test, padding], format='csr')
        print(f"Added {diff} zero-padding columns.")
    else:
        # Trim extra columns
        X_test = X_test[:, :expected_features]
        print(f"Trimmed {-diff} extra columns.")

# Predict
print("Predicting prices...")
test_preds = np.expm1(final_model.predict(X_test))
test["predicted_price"] = test_preds

# Save results
output_path = os.path.join(MODEL_DIR, "test_predictions.csv")
test[["sample_id", "predicted_price"]].to_csv(output_path, index=False)

print(f"Test predictions saved successfully â†’ {output_path}")
```

# Step 14 â€” Final Visualizations

```python
errors = y_val_pred - y_val_true
abs_errors = np.abs(errors)

plt.figure(figsize=(8,4))
sns.histplot(abs_errors, bins=50, kde=True, color='skyblue')
plt.title("Absolute Error Distribution")
plt.xlabel("Absolute Error (Predicted - Actual)")
plt.ylabel("Frequency")
plt.grid(alpha=0.3)
plt.show()

bins = [0, 100, 500, 1000, 5000, 10000, 50000, 100000, np.inf]
labels = ["<100", "100-500", "500-1k", "1k-5k", "5k-10k", "10k-50k", "50k-100k", ">100k"]
price_bins = pd.cut(y_val_true, bins=bins, labels=labels)

smape_by_range = []
for lbl in labels:
    idx = price_bins == lbl
    if idx.sum() > 0:
        smape_val = smape(y_val_true[idx], y_val_pred[idx])
        smape_by_range.append(smape_val)
    else:
        smape_by_range.append(np.nan)

plt.figure(figsize=(8,4))
sns.barplot(x=labels, y=smape_by_range, color='orange')
plt.title("SMAPE by Price Range")
plt.xlabel("Price Range")
plt.ylabel("SMAPE (%)")
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(7,5))
sns.scatterplot(x=y_val_true, y=errors, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Actual Price")
plt.ylabel("Prediction Error")
plt.title("Prediction Error vs Actual Price")
plt.show()

print("Visualization Summary:")
print(f"Mean Absolute Error: {np.mean(abs_errors):.2f}")
print(f"Median Absolute Error: {np.median(abs_errors):.2f}")
print(f"Overall SMAPE: {smape_score:.2f}%")
```
